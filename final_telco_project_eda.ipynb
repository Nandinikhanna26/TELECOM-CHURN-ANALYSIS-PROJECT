```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

```

```python
df = pd.read_csv('Cleaned_Customer_Data.csv')
```

```python
#Plot 1: Distribution of Monthly_Charge by Customer_Status
plt.figure(figsize=(10, 6))
sns.boxplot(x='Customer_Status', y='Monthly_Charge', data=df)
plt.title('Monthly Charge Distribution by Customer Status')
plt.xlabel('Customer Status')
plt.ylabel('Monthly Charge')
plt.savefig('monthly_charge_by_status.png')

plt.show()
```

```python
# Plot 2: Churn rate by Contract type
churn_rate = df.groupby('Contract')['Customer_Status'].value_counts(normalize=True).unstack().fillna(0)
churn_rate['Churn_Rate'] = churn_rate['Churned'] * 100
plt.figure(figsize=(8, 6))
sns.barplot(x=churn_rate.index, y=churn_rate['Churn_Rate'])
plt.title('Churn Rate by Contract Type')
plt.xlabel('Contract Type')
plt.ylabel('Churn Rate (%)')
plt.savefig('churn_rate_by_contract.png')

plt.show()
```

```python
# Plot 3: Correlation heatmap of numeric features
numeric_cols = ['Age', 'Number_of_Referrals', 'Tenure_in_Months', 'Monthly_Charge', 
                'Total_Charges', 'Total_Extra_Data_Charges', 'Total_Long_Distance_Charges']
plt.figure(figsize=(10, 8))
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Numeric Features')
plt.savefig('correlation_heatmap.png')
plt.show()
```

```python
# Plot 4: Churn distribution by Internet_Type
plt.figure(figsize=(10, 6))
sns.countplot(x='Internet_Type', hue='Customer_Status', data=df[df['Internet_Service'] == 1])
plt.title('Churn Distribution by Internet Type')
plt.xlabel('Internet Type')
plt.ylabel('Customer Count')
plt.savefig('churn_by_internet_type.png')
plt.show()
```
