```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r'/Users/nandinikhanna/Downloads/Data & Resources/Data/Customer_Data.csv')
```

```python
df
```

```python
# Handling negative Monthly_Charge values
median_charge = df[df['Monthly_Charge'] > 0]['Monthly_Charge'].median()
df['Monthly_Charge'] = df['Monthly_Charge'].apply(lambda x: median_charge if x <= 0 else x)
```

```python
df['Monthly_Charge'].
```

```python
# Replacing empty Value_Deal with "None"
df['Value_Deal'] = df['Value_Deal'].fillna('None')
```

```python
df['Value_Deal']
```

```python
#Checking for duplicates in Customer_ID
duplicate_ids = df[df['Customer_ID'].duplicated()]['Customer_ID']
if len(duplicate_ids) > 0:
    print(f"Found {len(duplicate_ids)} duplicate Customer_IDs: {duplicate_ids.tolist()}")
    df = df.drop_duplicates(subset='Customer_ID', keep='first')
```

```python
df['Customer_ID'].count()
```

```python
# Converting boolean-like columns to consistent format (Yes/No to 1/0)
boolean_columns = ['Married', 'Phone_Service', 'Multiple_Lines', 'Internet_Service', 
                  'Online_Security', 'Online_Backup', 'Device_Protection_Plan', 
                  'Premium_Support', 'Streaming_TV', 'Streaming_Movies', 
                  'Streaming_Music', 'Unlimited_Data', 'Paperless_Billing']
for col in boolean_columns:
    df[col] = df[col].map({'Yes': 1, 'No': 0, '': 0}).fillna(0).astype(int)
```
